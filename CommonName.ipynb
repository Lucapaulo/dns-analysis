```python
import numpy as np
import sqlite3 as sq
import matplotlib.pyplot as plt
import pandas as pd
import json
import tldextract

from colors import Colors

%matplotlib inline
```

```python
certs = pd.read_csv("certs.csv", converters={"info":json.loads}, header = None, names = ["ip", "protocol", "port", "info"])
```

```python
quic_certs = certs.loc[certs["protocol"] == "quic"]
quic_certs = quic_certs.groupby('ip', group_keys = False).apply(lambda df: df.sample(1))
# Expand common name
quic_certs["common_name"] = quic_certs["info"].apply(lambda info: tldextract.extract(info["CommonName"]).registered_domain)
quic_certs.groupby("common_name").size().reset_index(name = "size").nlargest(20, "size")
```
